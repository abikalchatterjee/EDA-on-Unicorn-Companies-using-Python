CONVERTING DATE JOINED FROM OBJECT to DATETIME

```python
companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])
companies.info()
```

CREATING YEAR JOINED COLUMN

```python
companies['Year Joined'] = companies['Date Joined'].dt.year
```

```python
companies.head(10)
```

CREATING A SAMPLE DATA

```python
companies_sampled = companies.sample(n = 50 , random_state= 42)
companies_sampled.head(10)
```

TIME TAKEN FOR COMPANY TO BECOME A UNICORN

```python
# Years it took to become a unicorn


companies_sampled['years_till_unicorn'] = companies_sampled['Year Joined'] - companies_sampled['Year Founded']

#Grouping data by industries

grouped = (companies_sampled[['Industry','years_till_unicorn']]
           .groupby('Industry')
           .max()
           .sort_values(by='years_till_unicorn')
           )
grouped
```

CREATING THE BAR PLOT

```python
#Creating th e bar plot using plt

plt.bar(grouped.index, grouped['years_till_unicorn'])

#Title

plt.title('Bar plot of maximum years taken by a company to become unicron per industry(sample)')

# x axis label

plt.xlabel('Industry')

# y axis label

plt.ylabel('Maximum number of years')

# Rotating label in the x axis to avoid overlap

plt.xticks(rotation= 45, horizontalalignment= 'right')

# Display the plot

plt.show
```

MAXIMUM UNICORN INDUSTRY VALUATION PER INDUSTRY

Converting the valuation from Object to int type for calcualtion

```python
# Creating new column valuation_billions

companies_sampled['valuation_billions'] = companies_sampled['Valuation']

#Removing $

companies_sampled['valuation_billions'] = companies_sampled['valuation_billions'].str.replace('$', '')


# Removing B representing billions

companies_sampled['valuation_billions'] = companies_sampled['valuation_billions'].str.replace('B', '')



```

```python
# Changing type to int

companies_sampled['valuation_billions'] = companies_sampled['valuation_billions'].astype('int')

companies_sampled.head()
```

Creating the grouped data fro plotting

```python
grouped_valuations = (companies_sampled[['Industry', 'valuation_billions']]
                      .groupby('Industry')
                      .max()
                      .sort_values(by='valuation_billions')
                      )
grouped_valuations
```

Plotting the bar plot


```python
#Plotiing the bar chart

plt.bar(grouped_valuations.index, grouped_valuations['valuation_billions'])


#Title for the bar plot


plt.title('Bar plot of maximum unicorn company valuation per industry (from sample')

# x axis label

plt.xlabel('Industry')

# y axis label


plt.ylabel('Maximun valuation in billions')

# Adjustiong the x axis label to prevent overlap

plt.xticks(rotation = 45, horizontalalignment= 'right')

# Showing the bar
 

plt.show()

```
